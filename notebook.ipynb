### **1)DATA CLEANING & PROCESSING**

*To prepare the data for modeling, we create a class called preprocessing that handles

```python
class DataPreprocessor:
    def __init__(self, df: pd.DataFrame):
        self.df = df.copy()

# remove missing values

    def handle_missing_values(self):
        """Fill or drop missing values (basic strategy)"""
# For now, we'll fill numeric columns with median
        self.df.fillna(self.df.median(numeric_only=True), inplace=True)
        print("[INFO] Missing values filled with column medians.")

# remove duplicates values

    def remove_duplicates(self):
        """Remove duplicate rows"""
        initial_shape = self.df.shape
        self.df.drop_duplicates(inplace=True)
        print(f"[INFO] Removed {initial_shape[0] - self.df.shape[0]} duplicate rows.")

#data processing

    def preprocess(self):
        """Run all preprocessing steps"""
        self.handle_missing_values()
        self.remove_duplicates()
        return self.df
```

### **Object creating**

```python

preprocessor = DataPreprocessor(df)
df = preprocessor.preprocess()    
print("\nNull Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

## **3) Bivariate Analysis Class (with return)**

### *What is Bivariate Analysis?*

**.** It examines the relationship between two variables, like:

**.** Correlation (for numerical-numerical)

**.** Crosstab (for categorical-categorical)

**.** Grouped statistics (categorical vs numerical)

```python
# Class to analyze data and create visual graphs

class UnivariateAnalysis:
    def __init__(self, df: pd.DataFrame, output_dir="graphs"):
        self.df = df  # Store the dataset
        self.output_dir = output_dir  # Folder to save graphs

        # Make the folder if it doesn't exist
        os.makedirs(self.output_dir, exist_ok=True)

    def plot_histograms(self):
        # Plot histograms for all numeric columns
        self.df.hist(bins=20, figsize=(14, 10))
        plt.suptitle("Distribution of Numerical Columns", fontsize=16)
        plt.tight_layout()

        # Save the histogram as an image
        save_path = os.path.join(self.output_dir, "histograms.png")
        plt.savefig(save_path)
        plt.close()
        print(f"[INFO] Histograms saved to {save_path}")

    def plot_heatmap(self):
        # Plot heatmap showing correlation between numeric columns
        plt.figure(figsize=(10, 8))
        sns.heatmap(self.df.corr(numeric_only=True), annot=True, cmap='coolwarm')
        plt.title("Correlation Heatmap")

        # Save the heatmap as an image
        save_path = os.path.join(self.output_dir, "heatmap.png")
        plt.savefig(save_path)
        plt.close()
        print(f"[INFO] Heatmap saved to {save_path}")
        
# Run both histogram and heatmap functions

    def run_all(self):
        self.plot_histograms()
        self.plot_heatmap()


# Use the class with your DataFrame

analyzer = UnivariateAnalysis(df)
analyzer.run_all()

# Show histograms on screen

df.hist(bins=20, figsize=(14, 10))
plt.suptitle("Distribution of Numerical Columns", fontsize=16)
plt.tight_layout()
plt.show()

# Show heatmap on screen
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### **Data Splitting (Train/Test Split)**

 *What is Data Splitting?*

**.** Splitting data into:

**.** Training set â used to train the model.

**.** Testing set â used to evaluate model performance.

**.** We typically use train_test_split() from sklearn.model_selection.

```python
from sklearn.model_selection import train_test_split
import pandas as pd

class DataSplitter:
    def __init__(self, df: pd.DataFrame, target_column: str, test_size=0.2, random_state=42):
        self.df = df
        self.target_column = target_column
        self.test_size = test_size
        self.random_state = random_state

    def split(self):
        X = self.df.drop(columns=[self.target_column])
        y = self.df[self.target_column]

        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=self.test_size, random_state=self.random_state
        )

        print(f"[INFO] Data split into train and test sets.")
        print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
        return X_train, X_test, y_train, y_test



splitter = DataSplitter(df, target_column="amount_paid")
X_train, X_test, y_train, y_test = splitter.split()
```
